```python
# リスト × 辞書（データ分析の基本形）

# 歩数と体重のデータ（辞書のリスト）
records = [
    {"date": "2025-08-01", "steps": 12000, "weight": 78.5},
    {"date": "2025-08-02", "steps": 15000, "weight": 78.25},
    {"date": "2025-08-03", "steps": 17769, "weight": 78.0}
]

# 一つずつ取り出す
for r in records:
     print(r["date"], r["steps"], r["weight"])

# 平均歩数の計算
total = 0
for r in records:
     total += r["steps"]

print("平均歩数：", total / len(records))
```

```python
# 複数の戻り値 × データ処理関数

def judge(steps):
    if steps >= 15000:
        return "excellent", steps * 0.04
    elif steps >= 10000:
        return "good", steps * 0.04
    else:
        return "bad", steps * 0.04

status, cal = judge(17769)
print(status, cal)
```

```python
# 辞書のリストを加工して「新しいリスト」を作る

# 元データ
records = [
    {"date": "2025-08-01", "steps": 12000, "weight": 78.5},
    {"date": "2025-08-02", "steps": 15000, "weight": 78.25},
    {"date": "2025-08-03", "steps": 17769, "weight": 78.0}
]

# 歩数に +1000 歩した新しいデータを作る
new_records = []

for r in records:
    new_r = {
        "date": r["date"],
        "steps": r["steps"] + 1000,
        "weight": r["weight"]
    }
    new_records.append(new_r)
    
print(new_records)
```
