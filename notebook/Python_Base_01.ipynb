```python
# リスト
scores = [85, 90, 78]

# 要素の取得
print(scores[0])
print(scores[1])

# 追加
scores.append(95)
print(scores)

# for と組合せ
for s in scores:
    print("スコア：", s)
```

```python
# 辞書

person = {
    "name": "Shuichi",
    "age": 61,
    "job": "Data Sience Learner"
}

print(person["name"])
print(person["age"])

# 値の更新
person["age"] = 62
print(person)

# for で辞書を回す
for key, value in person.items():
    print(key, "⇒", value)
```

```python
# 基本の関数
def greet(name):
    print(f"こんにちは、{name}さん！")
    
greet("シュウイチ")

# 足し算の関数
def add(a,b):
    return a + b

result = add(10, 20)
print("足し算の結果：", result)

#複数の戻り値
def get_status(weight):
    if weight < 60:
        return "やせ気味", weight
    else:
        return "標準～", weight

status, w = get_status(78)
print(status, w)
```
